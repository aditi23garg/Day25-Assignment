### Types of Machine Learning

Machine Learning is mainly divided into three types based on how the model learns from data:

### 1. Supervised Learning

In supervised learning, the model is trained using labeled data, which means both input and correct output are already given.
The model learns the relationship between input and output and then predicts results for new data.

Examples:

Email spam detection (Spam / Not Spam)
House price prediction

### 2. Unsupervised Learning

In unsupervised learning, the model works with unlabeled data, meaning no correct output is provided.
The model tries to find patterns, similarities, or groups in the data on its own.

Examples:

Customer segmentation
Product recommendation


### 3. Reinforcement Learning

In reinforcement learning, the model learns by interacting with the environment.
It takes actions and receives rewards or penalties based on those actions. Over time, it learns the best action to take.

Examples:

Robots learning to walk
Game playing (like chess or car racing)


```python
import pandas as pd

df = pd.DataFrame({
    'Age': [20, 25, 30],
    'Salary': [20000, 30000, 40000],
    'Bought': [0, 1, 1]})

X = df[['Age', 'Salary']]

y = df['Bought']

print(X)
print(y)
```

### Difference between Regression and Classification

Regression and classification are both types of supervised learning, but they are used for different types of problems.

### Regression

Regression is used when the output is a continuous value (number).
It predicts values that can be in a range.

Examples:

Predicting house prices
Predicting temperature

Explanation:
The output can be any number like 20, 35.5, 100, etc.

### Classification

Classification is used when the output is a category or class.
It predicts which group the data belongs to.

Examples:

Spam or Not Spam
Pass or Fail

Explanation:
The output is limited to specific categories.

### ✅ Types of Machine Learning with Real-World Examples

Machine Learning is a method where computers learn from data and make decisions without being directly programmed. There are three main types of machine learning:

### 🔹 1. Supervised Learning

Supervised learning is a type of machine learning where the model is trained using labeled data. This means the input data is already given along with the correct output.

The model learns the relationship between input and output and then uses this knowledge to predict results for new data.

👉 Real-world examples:

Email spam detection (Spam / Not Spam)
Predicting house prices based on area and location
Predicting student marks

👉 Key Point:
Data is already labeled, and the model learns from correct answers.

### 🔹 2. Unsupervised Learning

Unsupervised learning works with unlabeled data, meaning there is no predefined output. The model tries to find patterns, similarities, or groups in the data on its own.

👉 Real-world examples:

Customer segmentation in shopping websites
Grouping similar products in e-commerce
Detecting unusual patterns in bank transactions

👉 Key Point:
No labels are given; the model discovers hidden patterns.

### 🔹 3. Reinforcement Learning

Reinforcement learning is a type of machine learning where the model learns by interacting with its environment. It receives rewards for correct actions and penalties for wrong actions, and improves over time.

👉 Real-world examples:

Self-driving cars learning how to drive
Robots learning to walk
Game-playing AI (like chess or video games)

👉 Key Point:
Learning is based on rewards and penalties.
