### Q1. What is EDA (Exploratory Data Analysis)?

EDA (Exploratory Data Analysis) is the process of **analyzing and understanding a dataset before applying any machine learning model**.  
It helps us explore the data using **summary statistics and visualizations**.

### In EDA, we:

- Check the structure of data (rows and columns)  
- Identify missing values  
- Understand data types (numerical or categorical)  
- Analyze patterns, trends, and relationships  

###  Why EDA is Important:

- It helps us **clean the data** (handle missing or incorrect values)  
- It helps us **understand the data better**  
- It helps in **detecting outliers or unusual values**  
- It ensures we use the **right model and features**  

````python
avg_temp = data['temp'].mean()
result = data[data['temp'] > avg_temp]
result
````

 Insights from `describe()` Function

The `describe()` function in Pandas provides a **statistical summary of numerical data**.  
It helps us understand the **distribution and behavior** of the dataset.

From `describe()`, we get:

- **Count** → Number of non-missing values in each column  
- **Mean** → Average value of the data  
- **Standard Deviation (std)** → Shows how much the data varies  
- **Minimum (min)** → Smallest value  
- **Maximum (max)** → Largest value  


Quartiles (Data Distribution)

- **25% (Q1)** → Lower values  
- **50% (Median)** → Middle value  
- **75% (Q3)** → Higher values  


Conclusion

The `describe()` function helps us:
- Understand data distribution  
- Identify outliers  
- Analyze spread of data

### Exploratory Data Analysis (EDA) with Pandas

EDA is the process of understanding your dataset before modeling. Here are the key steps:


### 1. Load the Data

```python
import pandas as pd

df = pd.read_csv('data.csv')
```


### 2. First Look at the Data

```python
df.head()        # First 5 rows
df.tail()        # Last 5 rows
df.shape         # (rows, columns)
df.columns       # Column names
df.dtypes        # Data types of each column
```


### 3. Summary Statistics

```python
df.describe()               # Stats for numeric columns (mean, std, min, max, etc.)
df.describe(include='all')  # Includes categorical columns too
```

Example output for a `temp` column:

| | temp |
|--|------|
| count | 100.0 |
| mean | 36.6 |
| std | 5.2 |
| min | 20.0 |
| max | 42.0 |


### 4. Check for Missing Values

```python
df.isnull().sum()                        # Count of nulls per column
df.isnull().sum() / len(df) * 100        # Null percentage per column
```

Handle missing values:

```python
df.dropna()                              # Drop rows with nulls
df['col'].fillna(df['col'].mean())       # Fill nulls with mean
```


### 5. Check for Duplicates

```python
df.duplicated().sum()    # Count duplicate rows
df.drop_duplicates()     # Remove duplicates
```


### 6. Value Counts (for Categorical Columns)

```python
df['gender'].value_counts()                    # Frequency of each category
df['gender'].value_counts(normalize=True)      # As proportions
```

Example output:

```
Male      60
Female    40
Name: gender, dtype: int64
```


### 7. Correlation Analysis

```python
df.corr()                              # Correlation matrix
df['sales'].corr(df['price'])          # Correlation between 2 columns
```

Heatmap visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```


### 8. Distribution of Numeric Columns

```python
df['age'].hist(bins=20)        # Histogram
df['age'].plot(kind='box')     # Boxplot to detect outliers
```


### 9. Groupby Analysis

```python
df.groupby('city')['sales'].mean()       # Average sales per city
df.groupby('gender')['salary'].sum()     # Total salary by gender
```


### 10. Outlier Detection

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['salary'] < Q1 - 1.5 * IQR) | (df['salary'] > Q3 + 1.5 * IQR)]
```

These steps give you a solid understanding of your data before building any model!
